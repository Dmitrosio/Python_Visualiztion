```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('vgsales.csv')

#The Most Profitable Games of All Time
data.head(15)
```

```python
# all the best selling games of 2015-2020 on PC
data[(data['Year'] >= 2015) & (data['Platform'] == 'PC')].head(30)
```

```python
#the most popular genres based on global sales of all time
data.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(12).plot(kind='bar', title='Most popular genres based on global sales of all time')
```

```python
#Projection of sales of action games and sports genres over the yaers based on global sales
data[data['Genre'].isin(['Action', 'Sports'])].groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack().plot(title='Action and Sports Genres over the years based on global sales')
```

```python
#The top 10 best selling action games of all times based on global sales
data[data['Genre'] == 'Action'].groupby('Name')['Global_Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar', title='Top 10 best selling action games of all times based on global sales', color=['red', 'green', 'blue', 'yellow', 'purple', 'orange', 'pink', 'brown', 'black', 'grey'])
```
